# กราฟเปรียบเทียบมูลค่าส่งออกปี2560-2562 เดือน มกราคม - เมษายน

```python
import re
def fillNumber(input):
    pattern = re.split('\.',str(input))
    text = pattern[0]
    num = len(list(text))//3
    if len(list(text))==0:
        num=0
    elif (len(list(text))%3) ==0 and len(list(text))/3:
        num = num-1
    word = list(text)
    temp = len(list(text)) - num*3
    ans=''
    ans = ans.join(word[:temp])
    for i in range(num):
        ans = ans+','
        for j in range(3):
            ans = ans+word[temp+j+(i*3)][0]
    ans = ans+'.'+pattern[1]
    return ans
```

```python
#function นำ row ของcolumn แรก ถึง column สุดท้ายมาต่อกันเป็น column เดียว ชื่อ df['value']
def appenMonth(data):
    df = pd.DataFrame(columns=['value'])
    num = len(data)-len(data)//12*len(data)
    print(num)
    for i in range(len(data.columns)-1):
        for j in range(12):
            df.loc[(i*12)+(j)] = data.iloc[j,i]
    for i in range(len(data.iloc[:,len(data.columns)-1:len(data.columns)].dropna())):
        df.loc[((len(data.columns)-1)*12)+i] = data.iloc[i,len(data.columns)-2:len(data.columns)-1].values[0]
    return df
```
